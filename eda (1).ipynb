```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("res_dpre.csv")
```

```python
df
```

```python
df.info()
```

```python
df.drop(["Unnamed: 0"], axis =1 , inplace = True)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
image = sns.regplot(x='RATING',y='RunTime',data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
plt.title('audience rating vs RunTime duration')
plt.xlabel('audience rating')
plt.ylabel('RunTime duration')
plt.show()
```

```python
image.figure.savefig('vis.png')
```
